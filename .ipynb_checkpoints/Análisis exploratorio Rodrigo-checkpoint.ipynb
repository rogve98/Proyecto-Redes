```python
import pandas as pd
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
```

```python
archivo = '/home/francisco/Facultad de Ciencias/2023-1 Redes/datos_rodrigo.csv'
datos = pd.read_csv(archivo,
#             nrows = 5,
            )
```

```python
datos['Área Geográfica'].value_counts()
```

```python
datos['Id Sector Económico'].value_counts()
```

```python
coordenadas = datos[['Latitud', 'Longitud']].values
```

```python
N = len(coordenadas)
```

```python
M = np.zeros([N, N])
```

```python
for i in range(N):
    for j in range(N):
        x1,y1 = coordenadas[i]
        x2,y2 = coordenadas[j]
        M[i,j] = np.sqrt((x2-x1)**2 + (y2-y1)**2)
```

```python
plt.imshow(M)
plt.colorbar()
```

```python
plt.imshow(M < 0.002)
```

```python
for h in np.arange(0.001,0.002, .00001):
    A = (M < h)*1
    G = nx.from_numpy_array(A)
    if nx.is_connected(G):
        print(h)
        break
```

```python
nx.is_connected(G)
```

```python
len(G.edges)
```

```python
nx.draw(G, node_size = 50)
```

```python
color = datos['Id Sector Económico'].values
color = color/color.max()
```

```python
nx.draw(G, node_size = 50, node_color = color, cmap = 'jet')
```

```python
pos = {}

for i in datos.index:
    pos[i] = (datos.loc[i,'Longitud'], datos.loc[i,'Latitud'])
```

```python
plt.figure(figsize = [10,5])
nx.draw(G, node_size = 50, node_color = color, cmap = 'jet', pos = pos)
```
